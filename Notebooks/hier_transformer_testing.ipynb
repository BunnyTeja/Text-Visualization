### LOADING TRAINING DATA

```python
#for mnb model
X_train, X_test, Y_train, Y_test = model.get_data()
#for transformer
X_train_final, X_test_final, Y_train, Y_test = model.get_data(for_transformer = True)
```

### LOADING HIERARCHICAL TRANSFORMER MODEL

```python
hier_model = model.config_model()
print(hier_model.summary())
```

### PERFORMANCE METRICS FOR HIERARCHICAL TRANSFORMER AND OLD KITE MODEL

```python
# Performance measure of Transformer model
pred = hier_model.predict(X_test_final)
labels = pred.argmax(axis = 1)
print(model.classification_report(labels, Y_test))
```

```python
# Performance measure of old KITE model
Y_pred_nb = []
for doc in X_test:
    a = model.classify(doc)
    Y_pred_nb.append(model.np.argmax(a[0]))

print(model.classification_report(Y_pred_nb, Y_test))
```
